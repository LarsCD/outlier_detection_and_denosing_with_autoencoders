```python
# Standaard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import ingebouwde dataset
from gapminder import gapminder
```

```python
# Haal uit gapminder alle Eurpeanse landen
europa = gapminder[gapminder['continent'] == 'Europe']
europa
```

```python
# Uit het geheel van gapminder, groeperen op basis van continent en bekijk de populatie
gapminder.groupby('continent')['pop'].sum()
```

```python
# Groeperen eerst per continent en dan per jaar, dan bereken de bevolking 
gapminder.groupby(['continent','year'])['pop'].sum().reset_index().head(10)
```

```python
import ipywidgets as widgets
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Maak een widgit die alle numerieke waardes op aanvraag laat zien
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
x_dict = {a : a for a in gapminder.select_dtypes(include=numerics).columns}

def barchart(x):
    return sns.histplot(x = x, data = gapminder)

widgets.interact(barchart, x = x_dict)
```

```python
# Maak een samenvatting van aantal landen per continent
gapminder.groupby('continent').apply(lambda g: "Dit continent heeft " + str(g['country'].nunique()) + " landen")
```
