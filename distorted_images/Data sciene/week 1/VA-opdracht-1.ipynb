## Onderdeel 5

De broers hebben tijdens een aantal wedstrijden tegen elkaar gereden (niet uitsluitend in dezelfde rit). Ook deze data staat op BrightSpace, we laden de data in en berekenen het verschil van het onderlinge resultaat.

```python

import pandas as pd
import matplotlib.pyplot as plt
schaatstijden = pd.read_csv('/Users/jermaineveneman/Desktop/Schaatstijden500m verschil.csv')

schaatstijden['Verschil'] = schaatstijden['Ronald'] - schaatstijden['Michel']

plt.hist(schaatstijden['Verschil'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram van verschil tussen Ronald en Michel')
plt.xlabel('Verschil (seconden)')
plt.ylabel('Aantal wedstrijden')
plt.grid(axis='y', alpha=0.75)
plt.show()

ronald_mean = schaatstijden['Ronald'].mean()
michel_mean = schaatstijden['Michel'].mean()

ronald_std = schaatstijden['Ronald'].std()
michel_std = schaatstijden['Michel'].std()

print("\nGemiddelde tijd Ronald:", round(ronald_mean, 2))
print("Gemiddelde tijd Michel:", round(michel_mean, 2))
print("Spreiding Ronald (standaardafwijking):", round(ronald_std, 2))
print("Spreiding Michel (standaardafwijking):", round(michel_std, 2))

if ronald_mean < michel_mean:
    print("\nConclusie: Ronald is gemiddeld sneller.")
elif michel_mean < ronald_mean:
    print("\nConclusie: Michel is gemiddeld sneller.")
else:
    print("\nConclusie: Beide zijn gemiddeld even snel.")

if ronald_std < michel_std:
    print("Ronald is consistenter in zijn tijden.")
else:
    print("Michel is consistenter in zijn tijden.")
```
