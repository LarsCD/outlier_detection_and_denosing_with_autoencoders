### 3. 
**Selecteer van beide datasets alleen de kolommen 'sofifa_id', 'club' en 'overall'. Geef de kolommen
'overall' en 'club' in de dataset players_20 een nieuwe naam. Voeg de datasets players_19 en
players_20 samen middels een left_join. Doe hetzelfde met een inner_join, wat is het
verschil tussen de datasets?**

https://datalab01.ict.hva.nl/w/fD7W7PsfkpiKovcTidMCac

```python
# Rename columns
players19_select = players19.loc[:,['sofifa_id','overall','club']]
players20_select = (players20.
                    loc[:,['sofifa_id','overall','club']].
                    rename(columns={"overall": "overall_20", "club": "club_20"}))
players20_select.head(1)
```

```python
players_left = pd.merge(players19_select, players20_select, how = 'left', on = 'sofifa_id')
print(players_left.shape) #Left join heeft 17547 rijen (evenveel als players19)
players_left.head(2)
```

```python
players_inner = pd.merge(players19_select, players20_select, how = 'inner', on = 'sofifa_id')
print(players_inner.shape) #Inner join heeft 13043 rijen

players_inner.head(2)
```

**Hoeveel unieke voetballers zitten er in beide datasets?
Hoeveel voetballers komen in beide datasets voor?**

```python
players_outer = pd.merge(players19_select, players20_select, how = 'outer', on = 'sofifa_id')
print(f"Aantal unieke spelers totaal: {players_outer.sofifa_id.nunique()}")
```

### 4. 
**Gebruik de uitkomt van de left join en bepaal per speler de maximale overall ranking (dat is dus de overall ranking uit 2019 of uit 2020) en sla deze op in een nieuwe variabele best_overall**

*tip: gebruik numpy's maximum() function of pandas.DataFrame.max met axis = 1*

```python
players_left = players_left.assign(best_overall = lambda x: np.maximum(x.overall,x.overall_20))
players_left.head(10)
```

```python
players_left['best_overall2'] = players_left[["overall", "overall_20"]].max(axis = 1)
players_left.head(10)
```

## Spotify top 200 data
### 7. 
**Laad de dataset SpotifyNL in. Zet de dataset opvolgorde van nummer, artiest, datum en
totaal aantal streams.**

```python
spot = pd.read_csv('SpotifyNL.csv')
print(spot.shape)
```

```python
spot = spot.sort_values(['Track.Name','Artist','Date','StreamsTotal'])
spot.head(100)
```

```python
pd.isna(spot).sum()
# Met behulp van de url kunnen we misschien nog wat track namen en artiesten invullen
spot = spot.dropna()
```

### 8. 
**StreamsTotal bevat de cumulatieve som van alle views tot en met de betreffende dag. Bereken nu het aantal streams op alleen de aangegeven dag per nummer en artiest.**

*Tip: gebruik diff(). Hoe gaat je om met NA waardes? tip: fillna()*

```python
spot['StreamsDay'] = (spot.
          sort_values(['Track.Name','Artist','Date']).
          groupby(['Track.Name','Artist']).
          StreamsTotal.diff().
          fillna(spot.StreamsTotal))

# De eerste dag dat een nummer in de lijst staat geeft diff() een NA. Daar vullen we dus de waarde uit SteamsTotal in
# want op de eerste dag is die gelijk aan StreamsDay

spot.head(10)
```

```python
###Uitwerking zonder piping
spot = spot.sort_values(['Track.Name','Artist','Date'])
spot["Streamsday"] = spot.groupby(['Track.Name','Artist'])["StreamsTotal"].diff()
spot["Streamsday"] = spot["Streamsday"].fillna(spot["StreamsTotal"])    

spot.head()
```

### 9. 
**Bereken de positie in de top 200 o.b.v. aantal streams op die dag
Tip: zet de dataset op volgorde van dag en aantal streams.**

```python
spot['position'] = (spot.
                    sort_values(['Date','StreamsDay'], ascending = [True,False]).
                    groupby('Date')
                    .StreamsDay.rank(ascending = False))
spot.head()
```

### 10. 
**Bepaal de stijging of daling van elk nummer t.o.v. de dag ervoor. Is het nummer nieuw in de
top 200, bereken dan het verschil t.o.v. 201**

```python
# Let op! Een stijging in de ranking is eigenlijk een daling van het getal in kolom 'position'. 
# Vandaar een minnetje voor de hele code. 

spot['stijging'] = -(spot.
          sort_values(['Track.Name','Artist','Date']).
          groupby(['Track.Name','Artist']).
          position.diff().
          fillna(spot.position - 201))

# Nu vullen we de NA's t.g.v. de diff() met 201 - de positie van die dag. 

spot.head()
```

### 11. 
**Maak groepen van 10 op basis van de positie in de top 200**

*Tip: maak gebruik van de cut() functie*

```python
spot['group'] = pd.cut(spot.position, 20)
spot.head()
```
